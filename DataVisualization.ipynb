```python
import pandas as pd # for data processing
import seaborn as sb #Â for visualizations
import matplotlib.pyplot as plt # for visualizations
from plotnine import *
```

```python
data = pd.read_csv('/content/titanic_cleaned.csv')
```

```python
data.head()
```

1) Bar plot showing the count of survivors and non-survivos based on gender

```python
# Group the data by 'Sex' and 'Survived' columns and calculate the count of each group
data_grouped = data.groupby(['Sex', 'Survived']).size().reset_index(name='Counts')

# Create a bar plot using Seaborn's barplot function
sb.barplot(x='Sex', y='Counts', hue='Survived', data=data_grouped)

# Set the title and labels for the plot
plt.title('Survivors and Non-Survivors Based on Gender')
plt.xlabel('Gender')
plt.ylabel('Count')

# Display the plot
plt.show()
```

The resulting visualization is a bar plot that represents the count of survivors and non-survivors based on gender. The x-axis represents gender (male and female), the y-axis represents the count of individuals, and the bars are color-coded to represent the survival status (survived and non-survived).

The bar plot allows for a visual comparison between the counts of survivors and non-survivors within each gender category. Insights that can be derived from this plot include identifying any gender-based patterns in survival rates and comparing the overall number of survivors to non-survivors based on gender.

2) Histogram or density plot showing the distribution of passenger ages, separated by survivors and non-survivors.

```python
# This line creates a new figure with a specified size of 20 inches in width and 15 inches in height.
plt.figure(figsize=(20,15))

# This line creates a histogram plot using Seaborn (sb). It uses the 'Age' column from the 'data' DataFrame where the 'Survived' column has a value of 1 (indicating survival).
# The histogram is plotted without a kernel density estimate (KDE) and is colored green. The label 'Survived' is assigned to this plot.
sb.histplot(data=data[data['Survived'] == 1], x='Age', kde=False, color='green',  label='Survived')

# Plotting the non-survivors. This line creates another histogram plot for non-survivors.
# It uses the same 'Age' column from the 'data' DataFrame but filters the data for rows where the 'Survived' column has a value of 0 (indicating non-survival).
# The histogram is also plotted without a KDE and is colored red. The label 'Not Survived' is assigned to this plot.
sb.histplot(data=data[data['Survived'] == 0], x='Age', kde=False, color='red', label='Not Survived')

#  This line sets the title of the plot as 'Distribution of Passenger Ages, Separated by Survival'.
plt.title('Distribution of Passenger Ages, Separated by Survival')
# This line sets the label for the x-axis of the plot as 'Age'.
plt.xlabel('Age')
# This line sets the label for the y-axis of the plot as 'Count'.
plt.ylabel('Count')
# This line displays the legend on the plot, showing the labels for the different histogram plots.
plt.legend()
# This line displays the plot on the screen.
plt.show()
```

The resulting visualization is a histogram that represents the distribution of passenger ages, separated by survival status. The x-axis represents age, and the y-axis represents the count of passengers falling into each age group. The histogram is color-coded to differentiate between survivors (green) and non-survivors (red).

Insights that can be derived from this plot include understanding the age distribution of survivors and non-survivors. It allows for visual comparison of the count of survivors and non-survivors at different age ranges, which may reveal any age-related patterns in survival rates.

3) Box plot showing the distribution of fares based on passenger class.

```python
# # Mapping 'Pclass' to x-axis and 'Fare' to y-axis
p = (ggplot(data, aes(x='Pclass', y='Fare'))
     + geom_boxplot() # Creating a box plot based on the mappings
     + labs(title='Distribution of Fares Based on Passenger Class',
            x='Passenger Class',
            y='Fare') # Setting the title, x-axis label, and y-axis label

# Customizing the figure size
     + theme(figure_size=(20, 15)))

# Prints the plot object to display the plot.
print(p)
```

This code generates a box plot that visualizes the distribution of fares based on passenger class. The x-axis represents the passenger class, and the y-axis represents the fare. Each box plot displays the median, quartiles, and any outliers within each passenger class.

The title of the plot is set to "Distribution of Fares Based on Passenger Class", clearly describing the purpose of the visualization. The x-axis is labeled as "Passenger Class", while the y-axis is labeled as "Fare".

The geom_boxplot() function creates the box plots based on the provided aesthetic mappings. It automatically calculates the statistical summaries necessary to construct the box plots, such as the median, quartiles, and any outliers.

The theme(figure_size=(20, 15)) function customizes the figure size, setting it to 20 inches in width and 15 inches in height.

Overall, this visualization allows us to compare the distribution of fares across different passenger classes and identify any variations or outliers. It can provide insights into the fare differences among passenger classes and potentially indicate the presence of price disparities or unique fare patterns within each class.

4) Scatter plot showing the relationship between passenger age and fare, color-coded by survival status.

```python
# Mapping 'Age' to x-axis, 'Fare' to y-axis, and 'Survived' to color
p = (ggplot(data, aes(x='Age', y='Fare', color='factor(Survived)'))
     + geom_point() # Creating a scatter plot based on the mappings
     + labs(title='Relationship between Passenger Age and Fare, Colored by Survival Status',
            x='Age',
            y='Fare',
            color='Survived') # Setting the title, x-axis label, y-axis label, and color legend
     + theme(figure_size=(20, 15)))  # Customizing the figure size

# Print the plot
print(p)
```

This code generates a scatter plot that illustrates the relationship between passenger age and fare, with the points colored according to survival status.

The x-axis represents the passenger age, and the y-axis represents the fare. Each data point represents an individual passenger, with their age on the x-axis and fare on the y-axis.

The color of each point is determined by the survival status. The 'Survived' column is transformed into a factor using factor(Survived) to ensure proper color encoding. The different levels of survival (e.g., survived and not survived) will be assigned distinct colors in the plot.

The title of the plot is set to "Relationship between Passenger Age and Fare, Colored by Survival Status". The x-axis is labeled as "Age", the y-axis as "Fare", and the color legend as "Survived".

The geom_point() function creates the scatter plot based on the provided aesthetic mappings. Each data point is represented as a dot on the plot.

The theme(figure_size=(20, 15)) function customizes the figure size, setting it to 20 inches in width and 15 inches in height.

Overall, this visualization allows us to explore the relationship between passenger age and fare, while also considering their survival status. By coloring the points according to survival, we can potentially identify any patterns or trends related to age, fare, and survival. It can provide insights into whether there are any notable differences in fare distribution or age composition between survivors and non-survivors.
